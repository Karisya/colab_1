```python
import os
from google.colab import drive
from PIL import Image
import matplotlib.pyplot as plt

#  Google Drive
drive.mount('/content/drive')

#     Google Drive
image_folder = "/content/drive/My Drive/Colab Notebooks/img"
output_folder = "/content/drive/My Drive/Colab Notebooks/out_img"

#     
os.makedirs(output_folder, exist_ok=True)

#    ( PNG, JPG, JPEG)
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg'))][:5]

#    (450x500 )
new_size = (450, 500)

#    (x1, y1, x2, y2) -  
crop_box = (100, 100, 300, 400)  # (  x, y,   x, y)

#  
for image_name in image_files:
    img_path = os.path.join(image_folder, image_name)

    #  
    img = Image.open(img_path)

    #  
    img_resized = img.resize(new_size)

    #   
    cropped_img = img_resized.crop(crop_box)

    #      ( )
    img_without_crop = img_resized.copy()
    draw = Image.new("RGB", (crop_box[2] - crop_box[0], crop_box[3] - crop_box[1]), "white")
    img_without_crop.paste(draw, (crop_box[0], crop_box[1]))

    #   
    img_resized.save(os.path.join(output_folder, f"resized_{image_name}"))
    cropped_img.save(os.path.join(output_folder, f"cropped_{image_name}"))
    img_without_crop.save(os.path.join(output_folder, f"without_crop_{image_name}"))

    #  
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    axes[0].imshow(img_resized)
    axes[0].set_title("  (450x500)")
    axes[0].axis("off")

    axes[1].imshow(cropped_img)
    axes[1].set_title(" ")
    axes[1].axis("off")

    axes[2].imshow(img_without_crop)
    axes[2].set_title("   ")
    axes[2].axis("off")

    plt.show()

print(" !   :", output_folder)
```
